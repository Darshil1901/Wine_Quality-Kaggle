**Step 6: Analysis and Visualization**

```python
qc = df['quality'].value_counts()
plt.figure(figsize = (12,6))
plt.bar(qc.index, qc, color = "pink")
plt.title("Count Plot of quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.grid(axis = "y")
plt.show()
```

```python
#seaborn #when u have to visualize the category values then we will use the swarmplot 
plt.figure(figsize = (12,6))
sns.swarmplot(x = "quality", y = "alcohol", data = df, palette = "viridis")
plt.title("Count Plot of quality")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.grid(axis = "y")
plt.show()
```

This plot depicts thats the higher point density in specific regions shows the concentration indicating where the majority of the data points cluster. The points isolated and are far away from the clusters shows the outliers

```python
#seaborn #when u have to visualize the distrubution (multiple groups) numerical values then we will use the violinplot
df['quality'] = df['quality'].astype(str)
plt.figure(figsize = (12,6))
sns.violinplot(x = "quality", y = "alcohol", data = df, palette = {"3":"lightcoral", "4":"lightblue", "5":"lightgreen", "6":"gold", "7":"lightskyblue", "8":"lightpink"}, alpha = 0.9)
plt.title("Violin Plot for Quality and Alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.grid(axis = "y")
plt.show()
```

```python
#boxplot to summarize the visualization
plt.figure(figsize = (10,6))
sns.boxplot(x = "quality", y = "alcohol", data = df, palette = {"3":"lightcoral", "4":"lightblue", "5":"lightgreen", "6":"gold", "7":"lightskyblue", "8":"lightpink"})
plt.title("BoxPlot for Quality and Alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.grid(axis = "y")
plt.show()
```

```python
#relationshp between patterns and corelations we will use heat map
plt.figure(figsize = (15,10))
sns.heatmap(df.corr(), annot = True, fmt = '.2f', cmap = "Pastel1", linewidths = 2)
plt.title("Correlation HeatMap")
plt.show()
```

```python
#Univariate Analysis_1
sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))
plt.title('Univariate Analysis of volatile acidity', fontsize=20)
sns.histplot(data = df, x = 'volatile acidity', bins=30, kde=True, color='red')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
